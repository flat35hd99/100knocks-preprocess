> S-035: レシート明細データ（receipt）に対し、顧客ID（customer_id）ごとに売上金額（amount）を合計して全顧客の平均を求め、平均以上に買いをしている顧客を抽出し、10件表示せよ。ただし、顧客IDが"Z"から始まるものは非会員を表すため、除外して計算すること。

````python
%%sql
WITH customer_avg AS (
    SELECT customer_id,
        -- AVG(amount) AS avg_by_customer,
        AVG(amount) OVER (PARTITION BY customer_id) AS avg_by_customer,
        AVG(amount) OVER () AS avg_of_whole_customer
    FROM receipt
    WHERE customer_id NOT LIKE 'Z%'
    -- GROUP BY customer_id
)
SELECT customer_avg.customer_id, AVG(customer_avg.avg_by_customer) AS avg
FROM customer_avg
WHERE customer_avg.avg_by_customer >= customer_avg.avg_of_whole_customer
GROUP BY customer_avg.customer_id
LIMIT 10;
````

> S-038: 顧客データ（customer）とレシート明細データ（receipt）から、顧客ごとの売上金額合計を求め、10件表示せよ。ただし、売上実績がない顧客については売上金額を0として表示させること。また、顧客は性別コード（gender_cd）が女性（1）であるものを対象とし、非会員（顧客IDが"Z"から始まるもの）は除外すること。

> S-039: レシート明細データ（receipt）から、売上日数の多い顧客の上位20件を抽出したデータと、売上金額合計の多い顧客の上位20件を抽出したデータをそれぞれ作成し、さらにその2つを完全外部結合せよ。ただし、非会員（顧客IDが"Z"から始まるもの）は除外すること。

````python
%%sql
WITH top_customers_of_date AS (
    SELECT customer_id, COUNT(sales_ymd) AS count_sales_dates
    FROM receipt
    WHERE customer_id NOT LIKE 'Z%'
    GROUP BY customer_id
    ORDER BY count_sales_dates DESC
    LIMIT 20
    )
, top_customers_of_amount AS (
    SELECT customer_id, SUM(amount) AS sum_sales_amount
    FROM receipt
    WHERE customer_id NOT LIKE 'Z%'
    GROUP BY customer_id
    ORDER BY sum_sales_amount DESC
    LIMIT 20
    )
SELECT *
FROM top_customers_of_date
FULL OUTER JOIN top_customers_of_amount
    ON top_customers_of_date.customer_id = top_customers_of_amount.customer_id;
````

> S-041: レシート明細データ（receipt）の売上金額（amount）を日付（sales_ymd）ごとに集計し、前回売上があった日からの売上金額増減を計算せよ。そして結果を10件表示せよ。

````python
%%sql
WITH collect AS (
    SELECT SUM(amount), sales_ymd
    FROM receipt
    GROUP BY sales_ymd
    LIMIT 10
)
SELECT collect.sales_ymd, collect.sum - LAG(collect.sum, 1) OVER (ORDER BY collect.sales_ymd) AS diff
FROM collect
LIMIT 10;
````

> S-042: レシート明細データ（receipt）の売上金額（amount）を日付（sales_ymd）ごとに集計し、各日付のデータに対し、前回、前々回、3回前に売上があった日のデータを結合せよ。そして結果を10件表示せよ。

````python
%%sql
WITH collect AS (
    SELECT SUM(amount), sales_ymd
    FROM receipt
    GROUP BY sales_ymd
    LIMIT 10
)
SELECT
    collect.sales_ymd,
    collect.sum,
    LAG(collect.sum, 1) OVER (ORDER BY collect.sales_ymd) AS sales_yesterday,
    LAG(collect.sum, 2) OVER (ORDER BY collect.sales_ymd) AS sales_before_yesterday,
    LAG(collect.sum, 3) OVER (ORDER BY collect.sales_ymd) AS sales_before_before_yesterday
FROM collect
LIMIT 10;
````

````python
%%sql
WITH generation_customer AS (
    SELECT *, customer.age / 10 AS generation
    FROM customer
)
SELECT c.gender_cd, c.generation, SUM(r.amount)
FROM receipt AS r
    JOIN generation_customer AS c ON r.customer_id = c.customer_id
GROUP BY
    c.gender_cd, c.generation
````

>
> - 最小値以上第1四分位未満 ・・・ 1を付与
> - 第1四分位以上第2四分位未満 ・・・ 2を付与
> - 第2四分位以上第3四分位未満 ・・・ 3を付与
> - 第3四分位以上 ・・・ 4を付与

````python
%%sql
WITH sales_by_cid AS (
    SELECT customer_id, SUM(amount), PERCENT_RANK() OVER(ORDER BY SUM(amount)) AS percentile
    FROM receipt
    GROUP BY customer_id
)
SELECT sales.customer_id, sales.sum,
    CASE
        WHEN percentile >= 0.75 THEN 4
        WHEN percentile >= 0.5 THEN 3
        WHEN percentile >= 0.25 THEN 2
        ELSE 1
    END AS category
FROM sales_by_cid AS sales
LIMIT 10;
````
